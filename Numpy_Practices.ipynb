deep and shallow copy
deep copy and shallow copy describe how data is duplicated when creating a new variable from an existing object.

```python
Shallow Copy
#A shallow copy creates a new variable, but the data is still shared with the original object.
#If you modify the copied object, the original may also change.
```

```python
a = np.array([1,2,3,4])
b= a.view()  #shallow copy
b[0] = 99
print(a)
print(b)  #Here both arrays changed because they share the same memory.
```

```python
a = np.array([1,2,3,4])
b= a  #shallow copy
b[0] = 99
print(a)
print(b)
```

```python
a = ([1,2,3,4,5])
b = a.copy()
b[0]= 300
print(a)
print(b) #Now the original array remains unchanged
```

```python
#Shallow copy → shares memory with original
#Deep copy → creates independent memory
```
